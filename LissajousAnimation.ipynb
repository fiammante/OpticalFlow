# Create a Lissajous animation and save to mp4 video

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML

# Parameters for the Lissajous curve
A = 1  # Amplitude in x-direction
B = 1  # Amplitude in y-direction
a = 3  # Frequency in x-direction
b = 2  # Frequency in y-direction
delta = np.pi / 2  # Phase difference

# Time variable
t = np.linspace(0, 2 * np.pi, 1000)

# Create figure and axis
fig, ax = plt.subplots()
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
line, = ax.plot([], [], lw=2)

# Initialization function
def init():
    line.set_data([], [])
    return line,

# Animation function
def animate(i):
    y = A * np.sin(a * t + delta + i * 0.1)
    x = B * np.sin(b * t + i * 0.1)
    line.set_data(x, y)
    return line,

# Create animation
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=100, interval=20, blit=True)
FFwriter = animation.FFMpegWriter(fps=40)


plt.axis('off')
plt.grid(False)
plt.close()
ani.save('./lissajous.mp4', writer = FFwriter)
# Show the animation
HTML(ani.to_html5_video())
```
